```python
# MAAK EEN KOPIE VAN DE DATAFILE EN NOEM HEM ZO (HIER KOMT DE BINAIRE LABEL INFO BIJ): 
filepath = 'top1000_copy.dev'

# Read the existing content of the file
with open(filepath, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Modify each line by adding "0" at the end
modified_lines = []
for line in lines:
    q_id, doc_id, query, doc = line.split('\t')
    linez = line.strip()  # Remove any trailing newline characters
    if doc_id in rel_qid_did[q_id]:
        new_line = f"{linez}\t1\n"  # Append '1' separated by a tab and add a newline
    else:
        new_line = f"{linez}\t0\n"  # Append '0' separated by a tab and add a newline
    modified_lines.append(new_line)

# Write the modified lines back to the file
with open(filepath, 'w', encoding='utf-8') as file:
    file.writelines(modified_lines)
```
